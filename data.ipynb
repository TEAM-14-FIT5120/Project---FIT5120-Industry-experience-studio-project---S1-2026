```python
abs_stat_tan = pd.read_csv("Proportion of people 15 years and over who attempted suntanning(a) by state and territory, Nov 2023 to Feb 2024.csv")
```

```python
population_df = pd.read_csv("Annual population change at 30 June 2024.csv")
```

```python
population_df
```

```python
population_df = population_df[:-1]
```

```python
population_df
```

```python
population_df = population_df.drop(population_df.columns[[2,3]], axis=1)  # Drops the first 2 columns
```

```python
suntan_sunburn_df = pd.merge(abs_stat_sunburn, abs_stat_tan, on=['Region'], suffixes=('_sunburn', '_tan'))
```

```python
suntan_sunburn_df
```

```python
chart_2_df = pd.merge(suntan_sunburn_df, population_df, on=['Region'])
```

```python
chart_2_df
```

```python
chart_2_df["count_sunburn"] = chart_2_df["Proportion (%)_tan"] * chart_2_df["Population at 30 June 2024 ('000)"]*10
```

```python
chart_2_df
```

```python
chart_2_df.to_csv('chart_2_data.csv', index=False)
```
